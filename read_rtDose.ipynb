```python
import dicom as dcm
import SimpleITK as sitk

import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np

from __future__ import print_function
from ipywidgets import interact
```

```python
def ct_viewer(im):
    def plot_image(myslice):
        plt.imshow(im[myslice])#, cmap=plt.cm.gray)
    interact(plot_image, myslice=(0, im.shape[0]))
    return;
```

```python
rd=dcm.read_file('6E/EMC_LUNG_6E.dcm')
print rd
```

```python
rdpix=rd.pixel_array
rdim = sitk.GetImageFromArray(rdpix)

offset=[]
for val in rd.ImagePositionPatient: offset.append(float(val))
    
spacing=[]
for val in rd.PixelSpacing: spacing.append(float(val))   
spacing.append(float(rd.GridFrameOffsetVector[1]-rd.GridFrameOffsetVector[0]))

rdim.SetOrigin(offset)
rdim.SetSpacing(spacing)
sitk.WriteImage(rdim, 'test.mhd')
```

```python
im=sitk.ReadImage('6E/LUNG_6E.mhd')
impix=sitk.GetArrayFromImage(im)
ct_viewer(impix)
impix=impix[:, ::-1, :]
ct_viewer(impix)
impix=np.rot90(impix, k=1, axes=(0,1))
ct_viewer(impix)
```
